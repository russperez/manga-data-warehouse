```python
import pandas as pd
```

```python
try:
    df = pd.read_csv("mangadex_data.csv", encoding="utf-8")
except UnicodeDecodeError:
    df = pd.read_csv("mangadex_data.csv", encoding="ISO-8859-1")

print("File loaded! Shape:", df.shape)
df.head()
```

```python
# Drop rows
df.dropna(subset=["id", "title"], inplace=True)

# Ensure columns exist
for col in ["last_chapter", "year", "tags", "demographic", "status", "description"]:
    if col not in df.columns:
        df[col] = None  # Add the column with empty values if missing

# Converts 'last_chapter' and 'year' to integers, fill missing values with 0
df["last_chapter"] = pd.to_numeric(df["last_chapter"], errors="coerce").fillna(0).astype(int)
df["year"] = pd.to_numeric(df["year"], errors="coerce").fillna(0).astype(int)

# Normalize tags
df["tags"] = df["tags"].fillna("").apply(lambda x: [tag.strip() for tag in str(x).split(",") if tag.strip()])

# Clean up demographic and status values
df["demographic"] = df["demographic"].fillna("Unknown").astype(str).str.title()
df["status"] = df["status"].fillna("Unknown").astype(str).str.title()

# Clean text fields
df["description"] = df["description"].fillna("").astype(str)
df["title"] = df["title"].astype(str)

# Add description length
df["description_length"] = df["description"].apply(len)

# ccleaned data
print("Cleaned shape:", df.shape)
df.head()
```

```python
df.to_csv("mangadex_cleaned.csv", index=False, encoding="utf-8")
print("Cleaned data saved as 'mangadex_cleaned.csv'")
```
